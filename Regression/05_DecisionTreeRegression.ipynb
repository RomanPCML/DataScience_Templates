<h2> Decision Tree Regression</h2>

CARTs (Classification and Regression Trees) offer a good way for estimating values or classes especially if the data provides a larger number of independet variables. By splitting the dataset into subsets, as heterogen as possible until no further additional data can be created from splitting the subsets into further subsets. the data points to be predicted, can then follow the applied splitting rules until thes reach a final (terminal) leaf. The prediction will then be the average value of all data points in this leaf.

Further reading on [TowardsDataScience](https://towardsdatascience.com/decision-tree-algorithm-explained-83beb6e78ef4)


```python

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('../sample_data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor, plot_tree
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python

# Visualising the Decision Tree Regression results (higher resolution)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Because the Decision Tree Regressor is asigning the average value per leaf to the given data point. The regression graph shows a step wise ascend. The steps show the limits for each split in the decision tree and the horizonal lines are located at the class average.

```python
# Predicting a new result
y_pred = regressor.predict([[6.5]])
y_pred[0]
```
